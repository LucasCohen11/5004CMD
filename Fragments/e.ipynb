```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'../data/Trips_Full Data.csv')
```

```python
cols = [x for x in df.columns if 'Trips ' in x]
```

```python
data = df[cols].sum()
```

```python
data.sort_values().plot(kind="bar")
plt.title("Total amount of trips per range")
plt.show()
```
